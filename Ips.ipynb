## Genrating 500 IP Addresses

````python
import socket
import struct
import random

def generate_random_ip():
    return socket.inet_ntoa(struct.pack('>I', random.randint(1, 2**32 - 1)))

def generate_random_ip_addresses(num_addresses):
    return [generate_random_ip() for _ in range(num_addresses)]

if __name__ == "__main__":
    num_addresses = 500
    random_ip_addresses = generate_random_ip_addresses(num_addresses)
````

## Scrapping Data 

````python
# Defining Empty List
Country=[]
Decimal=[]
Hostname=[]
ISP=[]
Services=[]
State_Region=[]
City=[]
Latitude=[]
Longitude=[]

# Website URL
website_url = 'https://whatismyipaddress.com/'


# Webdriver Path
driver = webdriver.Chrome()

# Loop through each IP address
for ip_addres in random_ip_addresses[:50]:
    # Create a new instance of the webdriver for each iteration
    driver.get(website_url)

    # Find the search box element and enter the IP address
    search_box = driver.find_element(By.XPATH, "/html/body/div[1]/header/div[2]/form/label/input")
    search_box.clear()
    search_box.send_keys(ip_addres)

    # Find the search button element and click it
    search_button = driver.find_element(By.XPATH, "/html/body/div[1]/header/div[2]/form/button")
    search_button.click()

    # Wait for some time to allow the page to load
    time.sleep(3)

    
    try:
        # Country 
        country = driver.find_elements(By.XPATH, '/html/body/div[1]/div/div/div/div/article/div/div/div[1]/div/div[2]/div/div[1]/div/div[2]/div/div/div/div[1]/div[1]/p[7]/span[2]')
        # If data is found, extend with the text; otherwise, extend with "No country found"
        if country:
            Country.extend([name.text for name in country])
        else:
            Country.append("No country found")
            
        # Decimal
        decimal = driver.find_elements(By.XPATH, '/html/body/div[1]/div/div/div/div/article/div/div/div[1]/div/div[2]/div/div[1]/div/div[2]/div/div/div/div[1]/div[1]/p[1]/span[2]')
        # If data is found, extend with the text; otherwise, extend with "No decimal found"
        if decimal:
            Decimal.extend([name.text for name in decimal])
        else:
            Decimal.append("No decimal found")
            
        # Hostname
        hostname = driver.find_elements(By.XPATH, '/html/body/div[1]/div/div/div/div/article/div/div/div[1]/div/div[2]/div/div[1]/div/div[2]/div/div/div/div[1]/div[1]/p[2]/span[2]')
        # If data is found, extend with the text; otherwise, extend with "No hostname found"
        if hostname:
            Hostname.extend([name.text for name in hostname])
        else:
            Hostname.append("No hostname found")
            
        # ISP
        isp = driver.find_elements(By.XPATH, '/html/body/div[1]/div/div/div/div/article/div/div/div[1]/div/div[2]/div/div[1]/div/div[2]/div/div/div/div[1]/div[1]/p[3]/span[2]')
        # If data is found, extend with the text; otherwise, extend with "No ISP found"
        if isp:
            ISP.extend([name.text for name in isp])
        else:
            ISP.append("No ISP found")
            
        # Services
        services = driver.find_elements(By.XPATH, '/html/body/div[1]/div/div/div/div/article/div/div/div[1]/div/div[2]/div/div[1]/div/div[2]/div/div/div/div[1]/div[1]/p[4]/span[2]')
        # If data is found, extend with the text; otherwise, extend with "No services found"
        if services:
            Services.extend([name.text for name in services])
        else:
            Services.append("No services found")
            
        # state/region
        state_region = driver.find_elements(By.XPATH, '/html/body/div[1]/div/div/div/div/article/div/div/div[1]/div/div[2]/div/div[1]/div/div[2]/div/div/div/div[1]/div[1]/p[7]/span[2]')
        # If data is found, extend with the text; otherwise, extend with "No state/region found"
        if state_region:
            State_Region.extend([name.text for name in state_region])
        else:
            State_Region.append("No state/region found")
            
        # city
        city = driver.find_elements(By.XPATH, '/html/body/div[1]/div/div/div/div/article/div/div/div[1]/div/div[2]/div/div[1]/div/div[2]/div/div/div/div[1]/div[1]/p[8]/span[2]')
        # If data is found, extend with the text; otherwise, extend with "No city found"
        if city:
            City.extend([name.text for name in city])
        else:
            City.append("No city found")
            
        # Latitude
        latitude = driver.find_elements(By.XPATH, '/html/body/div[1]/div/div/div/div/article/div/div/div[1]/div/div[2]/div/div[1]/div/div[2]/div/div/div/div[1]/div[1]/p[10]/span[2]')
        # If data is found, extend with the text; otherwise, extend with "No Latitude found"
        if latitude:
            Latitude.extend([name.text for name in latitude])
        else:
            Latitude.append("No latitude found")
            
        # Longitude
        longitude = driver.find_elements(By.XPATH, '/html/body/div[1]/div/div/div/div/article/div/div/div[1]/div/div[2]/div/div[1]/div/div[2]/div/div/div/div[1]/div[1]/p[10]/span[2]')
        # If data is found, extend with the text; otherwise, extend with "No Longitude found"
        if longitude:
            Longitude.extend([name.text for name in longitude])
        else:
            Longitude.append("No longitude found")

        # print the title of the page
        print(f"Scraping data for IP address {ip_addres}: {driver.title}")

    except Exception as e:
            # Handle any exceptions that occur during scraping
        print(f"Error scraping data for IP address {ip_addres}: {str(e)}")

        

# Close the browser window
driver.quit()
````

````python
# Create a DataFrame
df = pd.DataFrame({
    'IP': random_ip_addresses[:50],
    'Country': Country,
    'Decimal': Decimal,
    'Hostname': Hostname,
    'ISP': ISP,
    'Services': Services,
    'State_Region': State_Region,
    'City': City,
    'Latitude': Latitude,
    'Longitude': Longitude
})
````
